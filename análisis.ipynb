### Importación de datos



```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación



Calcular el ingreso total de cada tienda, sumando los valores de la columna precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
ingreso_tienda1 = tienda1["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()

print(f"Ingreso total tienda 1: {ingreso_tienda1:,.2f}")
print(f"Ingreso total tienda 2: {ingreso_tienda2:,.2f}")
print(f"Ingreso total tienda 3: {ingreso_tienda3:,.2f}")
print(f"Ingreso total tienda 4: {ingreso_tienda4:,.2f}")
```

# 2. Ventas por categoría

calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
ventas_categoría_tienda1 = tienda1.groupby("Categoría del Producto").size()
ventas_categoría_tienda2 = tienda2.groupby("Categoría del Producto").size()
ventas_categoría_tienda3 = tienda3.groupby("Categoría del Producto").size()
ventas_categoría_tienda4 = tienda4.groupby("Categoría del Producto").size()

print("Tienda 1\n", ventas_categoría_tienda1)
print("\nTienda 2\n", ventas_categoría_tienda2)
print("\nTienda 3\n", ventas_categoría_tienda3)
print("\nTienda 4\n", ventas_categoría_tienda4)
```

# 3. Calificación promedio de la tienda


calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
promedio_tienda1 = tienda1["Calificación"].mean()
promedio_tienda2 = tienda2["Calificación"].mean()
promedio_tienda3 = tienda3["Calificación"].mean()
promedio_tienda4 = tienda4["Calificación"].mean()

print("Promedio Tienda 1:", round(promedio_tienda1, 2))
print("Promedio Tienda 2:", round(promedio_tienda2, 2))
print("Promedio Tienda 3:", round(promedio_tienda3, 2))
print("Promedio Tienda 4:", round(promedio_tienda4, 2))
```

# 4. Productos más y menos vendidos

identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
# Productos más y menos vendidos por tienda

# Tienda 1
mas_vendidos_tienda1 = tienda1["Producto"].value_counts().head(5).reset_index()
mas_vendidos_tienda1.columns = ["Producto", "Cantidad"]

menos_vendidos_tienda1 = tienda1["Producto"].value_counts().tail(5).reset_index()
menos_vendidos_tienda1.columns = ["Producto", "Cantidad"]

# Tienda 2
mas_vendidos_tienda2 = tienda2["Producto"].value_counts().head(5).reset_index()
mas_vendidos_tienda2.columns = ["Producto", "Cantidad"]

menos_vendidos_tienda2 = tienda2["Producto"].value_counts().tail(5).reset_index()
menos_vendidos_tienda2.columns = ["Producto", "Cantidad"]

# Tienda 3
mas_vendidos_tienda3 = tienda3["Producto"].value_counts().head(5).reset_index()
mas_vendidos_tienda3.columns = ["Producto", "Cantidad"]

menos_vendidos_tienda3 = tienda3["Producto"].value_counts().tail(5).reset_index()
menos_vendidos_tienda3.columns = ["Producto", "Cantidad"]

# Tienda 4
mas_vendidos_tienda4 = tienda4["Producto"].value_counts().head(5).reset_index()
mas_vendidos_tienda4.columns = ["Producto", "Cantidad"]

menos_vendidos_tienda4 = tienda4["Producto"].value_counts().tail(5).reset_index()
menos_vendidos_tienda4.columns = ["Producto", "Cantidad"]

# ==============================
# Mostrar productos más vendidos
# ==============================

print("=== PRODUCTOS MÁS VENDIDOS ===\n")

print("Tienda 1:\n", mas_vendidos_tienda1.to_string(index=False), "\n")
print("Tienda 2:\n", mas_vendidos_tienda2.to_string(index=False), "\n")
print("Tienda 3:\n", mas_vendidos_tienda3.to_string(index=False), "\n")
print("Tienda 4:\n", mas_vendidos_tienda4.to_string(index=False), "\n")

# ==============================
# Mostrar productos menos vendidos
# ==============================

print("=== PRODUCTOS MENOS VENDIDOS ===\n")

print("Tienda 1:\n", menos_vendidos_tienda1.to_string(index=False), "\n")
print("Tienda 2:\n", menos_vendidos_tienda2.to_string(index=False), "\n")
print("Tienda 3:\n", menos_vendidos_tienda3.to_string(index=False), "\n")
print("Tienda 4:\n", menos_vendidos_tienda4.to_string(index=False), "\n")
```

# 5. Envío promedio por tienda

calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.  Sabiendo que el cliente paga el envío.

```python
promedio_envio_tienda1 = tienda1["Costo de envío"].mean()
promedio_envio_tienda2 = tienda2["Costo de envío"].mean()
promedio_envio_tienda3 = tienda3["Costo de envío"].mean()
promedio_envio_tienda4 = tienda4["Costo de envío"].mean()

print(f"Costo de envío promedio tienda 1: {promedio_envio_tienda1:.2f}")
print(f"Costo de envío promedio tienda 2: {promedio_envio_tienda2:.2f}")
print(f"Costo de envío promedio tienda 3: {promedio_envio_tienda3:.2f}")
print(f"Costo de envío promedio tienda 4: {promedio_envio_tienda4:.2f}")
```
